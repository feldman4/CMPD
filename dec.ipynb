```python
class VersusDec(cmpd_web.Base):
    def reply(self, insult):

        self.insult += [insult['insult']]
        key = self.flag.next()
        self.wordbank = self.vocab[key]

        random.shuffle(self.wordbank)

        if key == self.vocab.keys()[0]:
            insult = ' '.join(self.insult)
            self.insult = []
        
            retort = dec_retort()
            score = self.score(insult)
            remark = {'insult': insult, 
                      'retort': retort,
                      'score': score}
                           
            self.emit('REMARK', remark)

            self.history += [remark]

        self.emit('UPDATE_WORDBANK', {'wordbank': self.wordbank})
```

```python
dec_vocab = cmpd_web.load_vocab('dec')
dec_keywords = [str(w.lower()) for s, words in dec_vocab[:2] for w in words]
adj = [str(w) for w in dec_vocab[0][1]]
noun = [str(w) for w in dec_vocab[1][1]]
```

```python
dec_text = open('resources/dec_all').read()
dec_text = dec_text.lower().translate(None, '?.",-()\xe2\x80\x9d\x9c:!\t;')
dec_text = dec_text.replace('\n', ' ')
dec_text = dec_text.replace('play boy', 'playboy')
open('resources/dec_text', 'w').write(dec_text)
```

```python
# bigram model
df_dec = pd.DataFrame(columns=dec_keywords, index=dec_keywords)
df_dec[:] = 0.0001

# deal with split words
tmp = str(dec_text)
for w in dec_keywords:
    tmp = tmp.replace(w, w.replace(' ', '_'))

tmp = tmp.split()
dec_bigrams = []
for w1, w2 in zip(tmp[:-1], tmp[1:]):
    w1 = w1.replace('_', ' ')
    w2 = w2.replace('_', ' ')
    if w1 in dec_keywords and w2 in dec_keywords:
        df_dec.loc[w1, w2] += 1

dec_counts = [[w, df_dec.loc[w].sum() + df_dec.loc[:,w].sum(), dec_text.split().count(w)] for w in dec_keywords]        

# normalize rows
df_dec = df_dec.div(df_dec.sum(axis=1), axis=0)
```

```python
def next_word(w):
    s = df_dec.loc[w]
    return s.sample(weights=s).index[0]
#     return s.sample().index[0]

def prev_word(w):
    s = df_dec.loc[:,w]
    return s.sample(weights=s).index[0]
```

```python
def make_phrase_fwd(start=None):
    if start:
        arr = [start]
    else:
        arr = [np.random.choice(adj)]
    while True:
        word = next_word(arr[-1])
        if word.lower() in noun:
            arr += [word]
            break
        elif word.lower() in arr:
            continue
        else:
            arr += [word]

    return ' '.join(arr).upper()

def make_phrase_rev():
    arr = [np.random.choice(noun)]
    while True:
        word = prev_word(arr[-1])
        if word.lower() in noun:
            break
        else:
            arr += [word]

    return ' '.join(arr[::-1]).upper()

dec_phrases = []
for a in adj:
    for _ in range(10):
        phrase = make_phrase_fwd(a)
        if len(phrase.split()) > 3:
            dec_phrases += [phrase]
            break
```
